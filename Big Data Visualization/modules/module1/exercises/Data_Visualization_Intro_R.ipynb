**Activity 4:** Fit a linear regression model between Calories and TotalFat.

**Activity 5:** Plot a grid of **scatter plots** for the 4 pairs of correlations identified in activity 3.
Make sure to add a **regression line**.

```r
## Your answer for activity 5 goes here ---

plot(USDA2$Calories, USDA2$TotalFat, xlab='Calories', ylab='TotalFat', col='blue')
abline(lm(USDA2$TotalFat ~ USDA2$Calories), col='red')

plot(USDA2$Calories, USDA2$SaturatedFat, xlab='Calories', ylab='SaturatedFat', col='blue')
abline(lm(USDA2$SaturatedFat ~ USDA2$Calories), col='red')

plot(USDA2$TotalFat, USDA2$SaturatedFat, xlab='TotalFat', ylab='SaturatedFat', col='blue')
abline(lm(USDA2$SaturatedFat ~ USDA2$TotalFat), col='red')

plot(USDA2$Carbohydrate, USDA2$Sugar, xlab='Carbohydrate', ylab='Sugar', col='blue')
abline(lm(USDA2$Sugar ~ USDA2$Carbohydrate), col='red')
```
